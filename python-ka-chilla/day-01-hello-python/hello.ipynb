### *Some Random Activity*

```python
iris_df = sns.load_dataset("iris")
iris_df.to_csv('./data/iris.csv', index=False)
print(iris_df.head())
```

```python
# Create a bar plot of the average sepal length for each species
plt.figure(figsize=(10, 6))
sns.barplot(x='species', y='sepal_length', data=iris_df, ci=None)
plt.title('Average Sepal Length for Each Species')
plt.xlabel('Species')
plt.ylabel('Average Sepal Length')
plt.show()
```

```python
# Pairplot to visualize relationships between variables
sns.pairplot(iris_df, hue='species')
plt.show()
```

```python
# Boxplot to visualize the distribution of each feature
plt.figure(figsize=(12, 8))
sns.boxplot(data=iris_df)
plt.title('Boxplot of Iris Dataset Features')
plt.show()
```

```python
# Violin plot to visualize the distribution and density of each feature
plt.figure(figsize=(12, 8))
sns.violinplot(x='species', y='sepal_length', data=iris_df)
plt.title('Violin Plot of Sepal Length by Species')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.violinplot(x='species', y='sepal_width', data=iris_df)
plt.title('Violin Plot of Sepal Width by Species')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.violinplot(x='species', y='petal_length', data=iris_df)
plt.title('Violin Plot of Petal Length by Species')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.violinplot(x='species', y='petal_width', data=iris_df)
plt.title('Violin Plot of Petal Width by Species')
plt.show()
```

```python
# Heatmap to visualize the correlation between features
plt.figure(figsize=(10, 6))
sns.heatmap(iris_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Iris Dataset Features')
plt.show()
```
